```python
#Outlier treatment (flooring, capping)
# taking backup before treating outliers
df_bef_OT=df.copy()
## Using User definition to return IQR, LR, UR for a column
def return_IQR_LR_UR(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=round((Q3-Q1),2)
    lower_range= round(Q1-(1.5 * IQR),2)
    upper_range= round(Q3+(1.5 * IQR),2)
    return IQR, lower_range, upper_range
```

```python
# creating a function to treate outlier and cap and floor upper and lower extreme values
def Outlier_treatment(df,num_cols):   
    i=0;
    for i in df.columns:
        if (i in num_cols):
            IQR, lr, ur = return_IQR_LR_UR(df[i])
            print ('col:', i , ',lower range :', lr, ',upper range:', ur, ',No of outliers:', df[(df[i]<lr)|(df[i]>ur)][i].count(), ',outlier %:', round(df[(df[i]<lr)|(df[i]>ur)][i].count()*100/df[i].count(),2))
            df[i] =np.where(df[i]>ur,ur,df[i])
            df[i] =np.where(df[i]<lr,lr,df[i])

    return df  
```

Skewed values in the dependent variable (target variable) can introduce several issues in statistical modeling and analysis:

1. **Biased Predictions:** especially when using models that assume a normal distribution of the target variable. Linear regression, for example, assumes normally distributed residuals, and if the target variable is highly skewed, the residuals may violate this assumption.

2. **Violated Assumptions:** Many statistical techniques, such as linear regression, assume that the residuals are normally distributed. Skewed distributions can result in non-normally distributed residuals, violating these assumptions and potentially leading to unreliable model results.

3. **Inflated Error Rates:** Skewed distributions can lead to overestimation or underestimation of errors, impacting the accuracy of the model. This can result in inflated error rates, making it challenging to assess the model's performance accurately.

4. **Impact on Significance Tests:** Skewed data can affect the results of hypothesis tests and statistical significance. If the dependent variable is not normally distributed, it may lead to inaccurate p-values and conclusions regarding the significance of predictor variables.

5. **Model Sensitivity:** Skewed distributions can make models sensitive to extreme values or outliers. This sensitivity can result in models being overly influenced by a few data points, leading to less robust and generalizable models.

6. **Difficulty in Interpretation:** Skewed data can make it challenging to interpret the relationship between the dependent variable and predictor variables accurately. It might be difficult to discern meaningful patterns or trends in the presence of skewness.

7. **Suboptimal Model Performance:** Skewed data can lead to suboptimal model performance, as models may struggle to capture the underlying patterns in the data. Transformations, such as log transformations, can help mitigate skewness and improve model performance.

Using log transformation, is one way to address skewed distributions, especially if the variable has a right-skewed (positively skewed) distribution. Log transformation can help stabilize the variance and make the distribution more symmetric, making it easier for the model to capture patterns in the data.

By default, the statsmodels library fits a line on the dataset which passes through the origin. But in order to have an intercept, you need to manually use the add_constant attribute of statsmodels. And once youâ€™ve added the constant to your X_train dataset, you can go ahead and fit a regression line using the OLS (Ordinary Least Squares) the attribute of statsmodels as shown below,

```python
# Add a constant to get an intercept
x_train_sm = sm.add_constant(x_train)
# Fit the resgression line using 'OLS'
model2 = sm.OLS(y_train, x_train_sm).fit()
```

```python
# Print the parameters,i.e. intercept and slope of the regression line obtained
print(model2.params)
print(model2.params.shape)
col2_sta = model2.params
```

```python
#Performing a summary operation lists out all different parameters of the regression line fitted
print(model2.summary()) #evaluation metrics
```
